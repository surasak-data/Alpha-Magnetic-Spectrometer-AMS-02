```python
def elecmodel(theta, E):
    C, alpha,beta,Eb , Ec = theta
    model = ( C * E**-alpha
            * (1+(E/Eb)**beta)**-1
            * np.exp(-E/Ec)*E**3 )
    return model

def posimodel(theta, E):
    Cs, alphas,beta, Ebs = theta
    model = ( Cs * E**-alphas
            * (1+(E/Ebs)**beta)**-1) *E**3
    return model

def source(theta, E):
    C2, alpha2, Ec2 = theta
    model = ( C2 * E**-alpha2
            * np.exp(-E/Ec2)*E**3 )
    return model

def total_background(elecmodel,posimodel,source): # 
    value =  elecmodel+ 1.6*posimodel+ 2*source # 
    return value

def background(elecmodel,posimodel):  
    value =  elecmodel+ 1.6*posimodel 
    return value
                  
```
